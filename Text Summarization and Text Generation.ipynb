**Hugging Face ile Otomatik Özetleme (Abstractive)**

```python
from transformers import pipeline


summarizer = pipeline("summarization", model="sshleifer/distilbart-cnn-12-6")


text = """
Artificial Intelligence (AI) is rapidly transforming the world. From healthcare to finance,
AI systems are being developed to solve complex problems and perform tasks that traditionally
required human intelligence. One of the most significant advances in AI has been the development
of deep learning algorithms that can learn from large datasets. These algorithms are behind
many of the recent breakthroughs in AI, such as self-driving cars, voice assistants, and
advanced medical diagnostics.
"""

summary = summarizer(text, max_length=60, min_length=30, do_sample=False)

print("Özet:")
print(summary[0]['summary_text'])
```

```python
from summa import summarizer

text = """
Yapay zeka, günümüzde birçok sektörde devrim yaratıyor. Sağlık, eğitim ve finans alanlarında
yapay zeka destekli sistemler kullanılarak verimlilik artırılıyor ve insan hataları azaltılıyor.
Özellikle makine öğrenmesi ve derin öğrenme algoritmaları, büyük veri setlerinden anlamlı sonuçlar
çıkararak yeni fırsatlar yaratıyor. Bu teknolojiler sayesinde gelecekte çok daha akıllı sistemler
görmemiz mümkün olacak.
"""


summary = summarizer.summarize(text, ratio=0.5)

print("Extractive Özet:")
print(summary)
```

**Hugging Face ile Basit Metin Üretimi (GPT-2)**

```python
from transformers import pipeline, set_seed

generator = pipeline('text-generation', model='gpt2')


prompt = "Artificial intelligence is"


set_seed(42)
output = generator(prompt, max_new_tokens=20, num_return_sequences=1)

print("Oluşturulan Metin:")
print(output[0]['generated_text'])
```

**Hugging Face ile Metin Üretimi (Model: GPT-Neo)**

```python
from transformers import pipeline, set_seed


generator = pipeline('text-generation', model='EleutherAI/gpt-neo-125M')


prompt = "The future of technology in education is"


set_seed(42)
output = generator(prompt, max_new_tokens=50, num_return_sequences=1)

print("Oluşturulan Metin:")
print(output[0]['generated_text'])
```
